Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a pandas DataFrame
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
# printing first 5 rows
credit_card_data.head()
```

```python
# printing last 5 rows
credit_card_data.tail()
```

```python
# dataset information
credit_card_data.info()
```

```python
# check for missing values in the dataset
credit_card_data.isnull().sum()
```

We do not have any missing values in this given dataset

```python
# checking the distribution of legit transactions and fraudulent transactions
credit_card_data['Class'].value_counts()
```

There are 2,84,315 legit transactions and 492 fraudulent transactions. We can conclude that this data is highly unbalanced.

Class value[0] -> legit transactions

Class value[1] -> fraudulent transactions

```python
# separating the data for analysis
normal = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(normal.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
normal.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# comparing the values for both transactions
credit_card_data.groupby('Class').mean()
```

Using these mean values, our model can check if the transaction is legit or fraud

Now we will perform under-sampling to deal with this unbalanced data.

We will build a sample dataset containing similar distribution of Normal and Fraudulent transactions

Number of Fraud transactions ==> 492

```python
# taking 492 datapoints randomly from normal transactions dataframe
normal_sample = normal.sample(n=492)
```

Concatenating this DataFrame with the Fraud transactions DataFrame

```python
new_dataset = pd.concat([normal_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

We performed the above steps to check whether we got a good sample or a bad one. In case of a bad sample, the mean values are very different.

Split the dataset into Features and Targets

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
print(X)
```

```python
print(Y)
```

Split the data into Training data and Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2) 
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training using Logistic Regression

```python
model = LogisticRegression() 
```

```python
# Training the Logistic Regression Model with the Training Data
model.fit(X_train, Y_train)
```

Model Evaluation

```python
# Accuracy Score

# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train) 
```

```python
print('Accuracy on Training data: ', training_data_accuracy)
```

Model Accuracy is approximately 94%. This means that out of 100 transactions, for almost 94 transacitions, the model will correctly predict if the transaction is Legit or Fraud. 

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data: ', test_data_accuracy)
```

Here, we have calculated accuracy scores on both Training data and Testing data. This is because, if the accuracy score on Training data is different from the accuracy score on Testing data, our model will be Underfitted or Overfitted. Since, here the accuracy scores are almost similar, we can conclude that our model performance is good.
